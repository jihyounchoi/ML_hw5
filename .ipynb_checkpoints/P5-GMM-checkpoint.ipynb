## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from matplotlib.colors import LogNorm
%matplotlib inline
```

## 1) 데이터 생성
랜덤하게 데이터를 생성하며 Trainset과 Testset으로 랜덤 샘플링하여 나누고 데이터셋이 어떤 분포로 생겼는지 시각화합니다.

```python
# 데이터 생성
X1, y1 = make_blobs(n_samples=1000, centers=((5, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.37, 0.95], [0.73, 0.6]]))
X2, y2 = make_blobs(n_samples=500, centers=1, random_state=42)
X = np.r_[X1, X2]

# 데이터를 훈련 데이터와 테스트 데이터로 분류
# X_test는 P5.B에서 사용
X_train, X_test = train_test_split(X, test_size=0.2)

# X_train 개수 : 1200
```

```python
# 훈련 데이터 plot
plt.plot(X_train[:, 0], X_train[:, 1], '.', markersize=3)
plt.title('Random Blobs for Training')
plt.show() 
```

```python
# 테스트 데이터 plot
plt.plot(X_test[:, 0], X_test[:, 1], '.', markersize=3)
plt.title('Random Blobs for Test')
plt.show() 
```

## 2) GMM 모델 구현, 학습, 시각화

Scikit-Learn의 GaussianMixture 이용하여 GMM을 학습해 봅니다.  
GaussianMixture의 `n_components`를 이용하면 군집의 갯수를 정할 수 있습니다. 이번 실습에서는 3개로 합니다.  
또한 GMM을 생성할 때 n_init을 10으로 설정해주세요

```python
gmm = GaussianMixture(n_components = 3, n_init = 10, covariance_type='full')    
# <your code> Gaussian Mixture 생성

gmm.fit(X_train)   
# <your code> 훈련 데이터(X_train)를 이용하여 학습
```

```python
def plot_gaussian_mixture(clusterer, X, show_centroid=True, show_contour=True, show_boundary=False):
    # clusterer ; GMM, X : X_train 입력
    
    # X는 여기에서는 X_train을 의미하며, axis = 0일 경우 각 열에 대한 최대/최소를 의미한다
    # 따라서 [input의 최대/최소, output의 최대/최소] 의 결과값이 리턴된다
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    
    
    x, y = np.meshgrid(np.linspace(mins[0], maxs[0], num=1000),np.linspace(mins[1], maxs[1], num=1000))
    # meshgrid : 좌표평면 상에 규칙적인 격자점을 생성하는 함수. x, y는 각각 격자점의 x좌표, y좌표를 의미하게 되며,
    # 순서를 맞추어 좌표점을 생성하면 grid 형태가 된다.
    # x, y는 각각 matrix 형태로 리턴됨
    
    print('x',len(x[0]))
    print('y',len(y[0]))
       
    # Centroids 시각화
    if show_centroid:
        
        
        centroids = clusterer.means_ # <your code> to use the mean of each components to get the centroids
        print('centroid : ', centroids)
        
        plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=20, linewidths=8, color='w', zorder=10)
        plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=20, linewidths=1, color='k', zorder=11) 
    
    # 밀도 등고선 시각화
    if show_contour:
        
        
        Z = (-1) * clusterer.score_samples(np.array([x.ravel(), y.ravel()]).T)
        # ravel은 주어진 matrix를 1차원 리스트로 변환한다.
        print('x : ', x)
        print('y : ', y)
        print('length z', len(Z))
        # <your code> to compute the negative log-likelihood of each sample
        
        # print(x.shape)
        Z = Z.reshape(x.shape) # x.shape : (1000, 1000)
        # Z의 길이는 1200인데, 이것을 1000 * 1000에 맞추어 reshape 하려니 오류가 발생한다.
        print(Z)
        
        
        cntr = plt.contourf(x, y, Z, norm=LogNorm(vmin=1.0, vmax=40), levels=np.logspace(0, 2, 10))
        plt.contour(x, y, Z, norm=LogNorm(vmin=1.0, vmax=40), levels=np.logspace(0, 2, 10), linewidths=1, colors='k')
    
    # 결정 경계 시각화
    if show_boundary:
        
        Z = clusterer.predict(np.array([x.ravel(), y.ravel()]).T)# <your code> to predict the labels
        Z = Z.reshape(x.shape)
        plt.contour(x, y, Z, linewidths=2, colors='r', linestyles='dashed')
    
    plt.xlabel("$f_1$")
    plt.ylabel("$f_2$", rotation=0)
```

```python
plt.figure(figsize=(7, 7))
plot_gaussian_mixture(gmm, X_train)
plt.plot(X_train[:, 0], X_train[:, 1], 'k.', markersize=4)
plt.show()
```

# Discussion

**1) 모수적 밀도추정과 비모수적 밀도추정을 비교하여 설명하세요.**

모수적 밀도추정이란, 데이터의 분포가 어떠한 형태의, 알려진 확률분포를 따른다는 가정에 기반한다. 따라서 어떠한 형태의 확률분포를 따른다고 가정한 후, 해당 확률분포의 parameter를 추정하는 방법을 모수적 밀도추정이라고 한다. 중심극한정리에 의해 모집단의 분포가 Gaussian(정규분포)를 따를 것이라고 가정하고 평균과 표준편차를 추정하는 사례가 대표적인 예시이다. 모수적 밀도추정은 비모수적 밀도추정에 비해 효율적이고, 일반적으로 높은 정확도를 지닌다. 하지만 데이터의 분포를 사전 지식을 통해 미리 가정하고 추정을 시작하므로, 모수적 가정이 잘못되었을 때 크게 잘못된 결론을 얻을 수 있는 위험성이 존재한다. 
이에 반해 비모수적 밀도추정은 데이터의 분포가 알려진 형태의 확률분포를 따르지 않는다고 가정하고 분포를 추정하는 방법이다. 비모수적 밀도추정은 일반적으로 모수적 밀도추정에 비해 비효율적이다. 하지만 데이터의 분포에 있어서 최소한의 가정만을 사용하므로, 모수적 가정이 잘못되어 생기는 오류의 가능성을 배제할 수 있다. Histogram을 이용하는 방법이나, Pazen Window 방법이 비모수적 밀도추정의 대표적인 예이다.


**2) EM 알고리즘을 이용한 GMM 밀도추정 방법에 대해  설명하세요.**

EM 알고리즘(Expectation-Maximization Algorithm) 이란 모수적 밀도추정의 일종이다. 특정 샘플의 소속정보 (여기서는 가우시안) 를 알기 위해서는 데이터들이 어떠한 분포를 따르는지 알아야 하고, 데이터들이 어떤 분포를 따르는지 알기 위해서는 샘플들의 소속정보를 알아야 하는, 물고 물리는 관계에 있다. 이러한 상태에서 데이터의 분포를 알아내기 위해, 우선 파라미터를 임의의 값으로 초기화한 후 해당 파라미터를 토대로 샘플의 소속정보를 추정한다 (Expectation). 그리고 구해진 소속정보를 토대로 파라미터를 업데이트하는 (Maximization) 일련의 과정을 반복하는 알고리즘을 EM이라고 한다. 이러한 알고리즘을 GMM에 적용하게 되면, EM을 통해 K개의 혼합 계수, K개의 평균벡터와 K개의 공분산행렬을 추정하게 된다.

우선 파라미터를 초기화하게 되는데, 파라미터의 초기화 방법에는 임의의 난수로 초기화하는 방법과, K-Means 알고리즘을 적용한 결과를 토대로 초기화하는 방법이 있다. 후자의 경우 평균벡터와 공분산행렬은 k번째 샘플들의 표본 - 평균벡터와 공분산행렬로, 혼합계수는 k번째 cluster에 속한 샘플의 수의 비율로 초기화하게 된다.

다음으로는 해당 파라미터 값을 토대로 샘플의 소속정보를 추정하게 된다. 이때 샘플들의 소속 정보는 행렬 z의 꼴로 나타내어지는데, z행렬의 ki 원소는 i번째 샘플이 k번째 가우시안에 속할 확률을 의미한다. 이는 k번째 가우시안의 posteriori probability를 의미하는데, 베이즈 정리를 통해 이 값을 구하여 소속정보를 업데이트한다.

업데이트된 소속정보를 토대로 다시 확률분포를 추정한다. 각 샘플들의 발생 확률이 서로 독립이라고 가정하며, 따라서 log-likelihood를 통해 log-확률 값들의 합으로 cost function을 정의할 수 있게 된다. 또한 모든 혼합 계수들의 합이 1이라는 제약조건을 함께 고려하게 되면, 결국 Lagrangian을 통해 해결할 수 있는 Constrained Optimization 문제가 된다. EM 알고리즘은 이러한 일련의 과정을 반복하는 방식으로 동작한다.

**3) K-means를 이용한 군집화와 GMM을 이용한 밀도추정 기반의 군집화의 차이에 대해 설명하세요.**

K-means와 GMM의 가장 큰 차이는 각 샘플이 어떠한 군집에 '확률적으로' 소속하는지, 그렇지 않은지의 차이에 있다.
K-means에서 특정 샘플은 특정 군집에 '속하거나 속하지 않는다'. 반면에 GMM은 특정 샘플이 특정 군집에 '특정 확률에 기반하여' 속하게 된다. 가령 샘플 x1과 군집 G1, G2가 존재한다고 하자. 이떄 K-means에서는 x1이 군집 G1에 소속하며, 군집 G2에는 소속하지 않는다고 말하게 된다. 하지만 GMM에서는 x1은 군집 G1에는 0.8, 군집 G2에는 0.2의 확률로 소속한다고 말하는 식이다. 이러한 차이로 인해 소속정보 행렬 z에도 차이가 있는데, K-means의 경우에는 이 행렬이 0 또는 1의 값만을 가지는 binary matrix가 되지만, GMM의 경우에는 각각의 값이 0 ~ 1 사이의 실수값을 가지게 된다.


**4) GMM의 Covariance type를 다른 type ('tied', 'spherical', 'diag')로 바꿀 수 있습니다. 또한, default Covariance type는 'full'로 클러스터의 모양, 크기, 방향에 제약이 없습니다.
그러면 Covariance type를 바꾸면 클러스터가 어떻게 바뀌는지 수행해보고, 이를 설명하세요.**

Covariance type이 'tied'인 경우, 생성되는 K개의 Gaussian이 동일한 모양을 가진다. 하지만 동일하게 가지는 모양의 종류에는 제한이 없다.
Covariance type이 'spherical'인 경우, 생성되는 K개의 Gaussian이 모두 원형의 (또는 구형의) 모양을 가진다.
Covariance type이 'diag'인 경우, 생성되는 K개의 Gaussian의 중심축이 좌표계의 축과 평행하는 형태를 가진다.
